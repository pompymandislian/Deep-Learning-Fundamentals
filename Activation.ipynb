## Activation Relu
---

When we must using this activation?

- When we want to change negative *values* to *zero*
- Avoid overfitting because this activation uses an important feature for computation, as the output is 0.
- Fast computation

Output:

- If the values input `x` are `positive` or `nol`, the output will be the `same` as the input; if they are `negative`, the output will be zero.

```python
display.Image("/content/drive/MyDrive/youtube/Deep Learning/Activation/Relu.png")
```

```python
# create a data for test
sample_relu = torch.tensor(range(-100, 50, 2))

sample_relu
```

```python
# scratch
def relu(x):
    """Activation relu"""
    y = torch.max(torch.tensor(0), x)
    return y

# let's try
relu(sample_relu)
```

```python
# sanicheck with torch module
torch.relu(sample_relu)
```

## Activation Tanh
---

When we must using this activation?
- When your data is centralized at 0, using tanh can spare data for -1 and 1.
- When you have a data complex
- When you need high expressiveness of the model (*ex: RNN*)

Output:

`-1` to `1`

```python
display.Image("/content/drive/MyDrive/youtube/Deep Learning/Activation/tanh.png")
```

- `e` : eksponen
- `z` : data input

```python
# scratch
def tanh(z):
  """Activation tanh"""
  x = torch.exp(z) - torch.exp(-z)
  y = torch.exp(z) + torch.exp(-z)

  result = x / y
  return result

# let's try
tanh(sample)
```

```python
# sanicheck with torch module
torch.tanh(sample)
```

## Activation SoftMax
---

```python
display.Image("/content/drive/MyDrive/youtube/Deep Learning/Activation/softmax.jpg")
```

When we must using this activation?

- Want to obtain probability with multiclass

Output :

`0` to `1`

```python
# create a data for test
sample_softmax = torch.tensor([1.0, 2.0, 3.0])
sample_softmax
```

```python
# scratch
def softmax(x):
    """Activation softmax"""
    exp_x = torch.exp(x)
    sum_exp_x = torch.sum(exp_x)
    result = exp_x / sum_exp_x

    return result

# let's try
softmax(sample_softmax)
```

```python
# sanicheck with torch module
torch.softmax(sample_softmax, dim = 0)
```

## Activation Leaky ReLU (Rectified Linear Unit)
---

When we must using this activation?

- When you obtain a dead neuron or 0 output in Relu during backpropagation
- Have small information

Output:
- If the values input x are positive or nol, the output will be the same as the input; if they are negative, then calculate small parameter with `  `

```python
display.Image("/content/drive/MyDrive/youtube/Deep Learning/Activation/leaky relu.png")
```

Formula:

$$ {Leaky Relu} = {max(0.01 * x, x)}$$

```python
# create Leaky Relu
import torch.nn as nn

class LeakyReLU(nn.Module):
    def __init__(self, negative_slope=0.01, inplace=False):
        """Define negative slope"""
        super(LeakyReLU, self).__init__()
        self.negative_slope = negative_slope
        self.inplace = inplace

    def forward(self, x):
        """Obtain leaky relu from module nn"""
        return torch.nn.functional.leaky_relu(x,
                                              negative_slope=self.negative_slope,
                                              inplace=self.inplace)
```

```python
# let's try

# instance
leaky_relu = nn.LeakyReLU(negative_slope=0.1)

# definition data
input_tensor = torch.randn(100)
leaky_relu(input_tensor)
```

# Simple Predicion Classification
---

Collect Data

```python
# generate data
import random

# random
purchase_price = [random.randint(0, 1000000) for rows in range(100)]
impression = [random.randint(0, 300) for rows in range(100)]
buy = [random.choice([True, False]) for rows in range(100)]

print('purchase_price', purchase_price[:5])
print('impression', impression[:5])
print('buy', buy[:5])
```

```python
import pandas as pd

# change to dataframe
data = {'purchase_price' : purchase_price,
        'impression' : impression,
        'buy' : buy}

# dataframe
df = pd.DataFrame(data)

df.head()
```

Preprocessing

```python
# change to label encoder

df['buy'] = df['buy'].apply(lambda x:1 if x else 0)
```

```python
df.head()
```

```python
# split data
from sklearn.model_selection import train_test_split

x = df[['purchase_price', 'impression']]
y = df['buy']

# split to train and test
X_train, X_test, y_train, y_test = train_test_split(x,
                                                    y,
                                                    test_size = 0.2,
                                                    random_state = 42,
                                                    stratify = y)
```

```python
# shape
print('X_train', X_train.shape)
print('y_train', y_train.shape)
print('X_test', X_test.shape)
print('y_test', y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

def Standartscaller(X, scaler=None):
    """Feature Engineering"""
    # initialize none
    scaled_data = None

    # scaling
    if scaler is None:
        scaler = StandardScaler()

        # fit data
        scaler.fit(X)

    # transform
    scaled_data = scaler.transform(X)

    return scaled_data, scaler

# scaling data
X_train, scaler = Standartscaller(X_train)
X_test, _ = Standartscaller(X_test, scaler)
```

```python
print('X_train after scaling', X_train[:5])
```

```python
print('X_test after scaling', X_test[:5])
```

```python
display.Image("/content/drive/MyDrive/youtube/Deep Learning/Activation/cross entropy.png")
```

- N is sum of sample in dataset
-  is true label (0 atau 1) for sampel i.
- ^i is prediction probabilty by model for sample i, probability class positive or negative


Why using cross entropy?

- Minimizing the difference between the probabilities predicted by the model and the true labels leads to a smaller loss value, indicating that the model better captures the observed probability distribution of the true labels.

```python
# preaper for output

def sigmoid(x):
    """Activation sigmoid"""
    output = 1 / (1 + torch.exp(-x))
    return output

def cross_entropy(y_prediction, y_true):
    """Find loss function for classification"""

    # Get the number of data points
    n = len(y_prediction)

    # Calculate the cross-entropy loss
    x = torch.log(y_prediction) * y_true
    y = (1 - y_true) * torch.log(1 - y_prediction)
    ce = -1/n * torch.sum(x + y)

    return ce
```

```python
display.Image("/content/drive/MyDrive/youtube/Deep Learning/Activation/flow.png")
```

```python
# parameter model
input_dim = 2
hidden_dim1 = 3
hidden_dim2 = 3
```

```python
import torch.nn as nn

class NeuralNetwork(nn.Module):
  def __init__(self, input_dim, hidden_dim1, hidden_dim2):
    super(NeuralNetwork, self).__init__()
    self.fc1 = nn.Linear(input_dim, hidden_dim1) # input layers
    self.fc2 = nn.Linear(hidden_dim1, hidden_dim2) # hidden layers
    self.fc3 = nn.Linear(hidden_dim2, 1) # output layers

  def forward(self, x):
      """Forward Loss Function"""
      x = sigmoid(self.fc1(x)) # activation input
      x = sigmoid(self.fc2(x)) # activation hidden
      x = sigmoid(self.fc3(x))  # activation output
      return x
```

```python
# create a model
model = NeuralNetwork(input_dim, hidden_dim1, hidden_dim2)
model
```

```python
# parameter training
num_epochs = 20
learning_rate = 0.01

# Initialize losses
train_losses = []
test_losses = []

# Training loop
for epoch in range(num_epochs):

    # Forward pass for training data
    outputs_train = model(torch.tensor(X_train, dtype=torch.float32))
    loss_train = cross_entropy(outputs_train,
                               torch.tensor(y_train, dtype=torch.float32))
    train_losses.append(loss_train.item())

    # Forward pass for testing data
    outputs_test = model(torch.tensor(X_test, dtype=torch.float32))
    loss_test = cross_entropy(outputs_test,
                              torch.tensor(y_test.values, dtype=torch.float32))
    test_losses.append(loss_test.item())

    # Print progress
    print(f'Epoch [{epoch+1}/{num_epochs}], Train Loss: {loss_train.item():.4f}, Test Loss: {loss_test.item():.4f}')

    model.zero_grad()  # Reset gradients
    loss_train.backward()  # Compute gradients

    # Manual adjustment of weights
    with torch.no_grad():
        for param in model.parameters():

            # Update weights using gradient descent
            param -= learning_rate * param.grad
```

```python
# Plot the losses
import matplotlib.pyplot as plt

plt.plot(range(1, len(train_losses)+1), train_losses,
         marker='o', linestyle='-', label = 'Train Loss')
plt.plot(range(1, len(test_losses)+1), test_losses,
         marker='o', linestyle='-', label = 'Test Loss')

plt.title('Loss vs Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.show()
```

So bad, because we not setting optimizer and other parameter, Just practice about activation

```python
# Evaluating the trained model
with torch.no_grad():
    y_pred_train = model(torch.tensor(X_train, dtype=torch.float32))

print('Predictions Train:', y_pred_train[:5])
```

```python
# Evaluating the trained model
with torch.no_grad():
    y_pred_test = model(torch.tensor(X_test, dtype=torch.float32))

print('Predictions Test:', y_pred_test[:5])
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

# Convert predictions to binary labels
y_pred_train = (y_pred_train > 0.5).float().numpy()

# Convert y_test to numpy array
y_train_numpy = y_train.values

# Calculate accuracy
accuracy = accuracy_score(y_train_numpy, y_pred_train)

print(f'Data Train Accuracy: {accuracy:.4f}')
```

```python
# Convert predictions to binary labels
y_pred_test = (y_pred_test > 0.5)

# Convert y_test to numpy array
y_train_numpy = y_test.values

# Calculate accuracy
accuracy = accuracy_score(y_train_numpy, y_pred_test)

print(f'Data Test Accuracy: {accuracy:.4f}')
```

```python
# confusion_matrix test
confusion_matrix(y_true = y_test,
                 y_pred = y_pred_test)
```
